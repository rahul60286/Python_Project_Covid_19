```python
covid_df.drop(["Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational"],inplace=True, axis=1)
```

```python
covid_df.head()
```

```python
covid_df['Date']= pd.to_datetime(covid_df['Date'], format = '%Y-%m-%d')
```

```python
covid_df.head()
```

```python
# active casses 
covid_df['Active_casses'] = covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])
```

```python
covid_df.tail()
```

```python
statewise = pd.pivot_table(covid_df, values =["Confirmed","Deaths","Cured"], index ="State/UnionTerritory", aggfunc = max)
```

```python
statewise["Recovery Rate"] = statewise["Cured"]*100 /statewise["Confirmed"]
```

```python
statewise["Mortality Rate"] = statewise["Deaths"]*100 /statewise["Confirmed"]
```

```python
statewise =statewise.sort_values(by ="Confirmed", ascending = False)
```

```python
statewise.style.background_gradient(cmap = "cubehelix")
```

```python
# Top 10 active cases

Top_10_active_cases =covid_df.groupby(by ='State/UnionTerritory').max()[['Active_casses','Date']].sort_values(by =['Active_casses'],ascending = False).reset_index()
```

```python
fig = plt.figure(figsize=(16,9))
```

```python
plt.title("Top 10 states with most Active Cases in India", size= 25)
```

```python
ax = sns.barplot(data = Top_10_active_cases.iloc[:10], y= "Active_casses", x="State/UnionTerritory", linewidth = 2, edgecolor = "red" )
```

```python
# Top 10 active cases

Top_10_active_cases =covid_df.groupby(by ='State/UnionTerritory').max()[['Active_casses','Date']].sort_values(by =['Active_casses'],ascending = False).reset_index()
fig = plt.figure(figsize=(16,9))
plt.title("Top 10 states with most Active Cases in India", size= 25)
ax = sns.barplot(data = Top_10_active_cases.iloc[:10], y= "Active_casses", x="State/UnionTerritory", linewidth = 2, edgecolor = "red" )

plt.xlabel("States")
plt.ylabel("Total Active Caces")
plt.show()
```

```python
# Top 10 States with highest Death rate 

top_10_deaths = covid_df.groupby(by ='State/UnionTerritory').max()[['Deaths','Date']].sort_values(by = ['Deaths'],ascending = False).reset_index()

fig = plt.figure(figsize=(18,5))

plt.title("Top 10 States with most Deaths ", size =25)

ax = sns.barplot(data = top_10_deaths.iloc[:12], y= "Deaths", x="State/UnionTerritory", linewidth = 2, edgecolor ="black")

plt.xlabel("States")
plt.ylabel("Total Death Case")
plt.show()
```

```python
# Growth Trends

fig = plt.figure(figsize = (12,6))

ax = sns.lineplot(data =covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra','Karnataka','Kerala','Tamil Nadu','Uttar Pradesh'])],x='Date', y='Active_casses', hue ='State/UnionTerritory')

ax.set_title("Top 5 Affected States In India", size=16)
```
